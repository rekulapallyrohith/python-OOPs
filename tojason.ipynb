```python
import yaml

ticket = {
    'theatre':'Gokul',
    'movie':'Salar',
    'seat':'A11',
    'date':'07-Nov-2025',
    'showtime':'11:30'
}
```

```python
with open('ticket.yaml', mode="w") as f:
    yaml.dump(ticket, f)
```

```python
with open('ticket.yaml', mode="r") as f:
    read_ticket = yaml.load(f, Loader=yaml.FullLoader)
```

```python
read_ticket
```

```python
with open('avengers.yaml', mode="r") as f:
    avengers_dict = yaml.safe_load(f)
```

```python
avengers_dict['movie']['distribution']['distributor']
```
